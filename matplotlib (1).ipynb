```python
#hist graphs are barcharts that are combined together 
#bins can be used to subgroup the values
#color is used 
#edgecolor are also used
#cumulative function is used to combine all the values the last part of the graph contains all the values before it 
```

```python
ages = np.random.normal(25,2,1000)
```

```python
plt.hist(ages,bins=[ages.min(),24,27,ages.max()],color='r',edgecolor='green',cumulative=True)
```

```python
#piechart
#explode moves single or more values apart from pie so can be distinguished easier 
#for explode amount that the element needed to seperate from rest is to be given in a list form [] for each element value
#autopct shows the amount of values that a element covers in percentage form
#pctdistance moves the percentage value from middle to out side the pie chart of easy viewing
#start angle helps to start piechart at specific angle
```

```python

langs =['python','c++','c','java','c#','Go']
votes =[25,14,12,34,13,3]
```

```python
plt.pie(votes, labels=langs)
plt.show()
```

```python
explodes =[0,0,0,0.1,0,0]
plt.pie(votes, labels=langs,explode=explodes,autopct='%.2f%%',pctdistance=1.3,startangle=90)
plt.show()
```

```python
#supplot with axis
x =np.arange(100)
fig, axs = plt.subplots(2,2)
axs[0,0].plot(x,np.sin(x))
axs[0,0].set_title('sine wave')
axs[0,1].plot(x,np.cos(x))
axs[0,1].set_title('cos wave')
axs[1,0].plot(x,np.log(x))
axs[1,0].set_title('log functon')
axs[1,1].plot(x,np.random.random(100))
axs[1,1].set_title('random function')
plt.suptitle('Fourplots')
plt.show()
```

```python
#3d plotting
#can lable x,y,z
```

```python
ax = plt.axes(projection='3d')
x =np.arange(0,50,0.1)
y =np.arange(0,50,0.1)
z =np.cos(x+y)
ax.set_title('3d projection')
ax.plot(x,y,z)
plt.show()
```

```python
#surface plot
#surface plot need a mesh type value or 2d array type values to plot
```

```python
x = np.arange(-5,5,0.1)
y = np.arange(-5,5,0.1)
X,Y =np.meshgrid(x,y)
Z =np.sin(X) * np.cos(Y)
ax.plot_surface(X,Y,Z, cmap='Spectral')
ax.set_title('3d plot')
ax.set_xlabel('test')
plt.show()
```

```python
#Animation 
import random
heads_tails =[0,0]
for _ in range(100000):
    heads_tails[random.randint(0,1)] += 1
    plt.bar(['Heads','Tails'],heads_tails, color=['red','blue'])
    plt.pause(0.001)
plt.show()
```
